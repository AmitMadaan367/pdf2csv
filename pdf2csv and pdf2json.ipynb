# To read the pdf file as dataframe

```python
import tabula

# add the path of your pdf file

pdf_path = 'FRANKIE_PDF/BROOKLYN COLLEGE-TRANS.pdf'
                                              
#read the pdf file

df=tabula.read_pdf(pdf_path, stream=True,pages='all')
```

# To convert pdf to csv file

```python
# add the name of pdf with path that you want to convert into csv file
# also give the name and path of generated csv file

tabula.convert_into('FRANKIE_PDF/BROOKLYN COLLEGE-TRANS.pdf', ("FRANKIE_PDF/BROOKLYN COLLEGE-TRANS.pdf2" + '.csv'), 
                    output_format="csv", pages="all")
print("csv file is generated")
```
